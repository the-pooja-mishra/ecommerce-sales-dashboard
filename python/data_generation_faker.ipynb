```python
from faker import Faker
import pandas as pd
import random

fake= Faker()

customers= []

for i in range(1, 1001):
    customers.append([i, fake.name(),
                      random.choice(['Male', 'Female']),
                      random.randint(18,65),
                      fake.city(),
                      fake.date_between(start_date='-3y', end_date='today')])

customers_df= pd.DataFrame(customers, columns=[ 'customer_id','customer_name','gender','age','city','signup_date'])

customers_df.to_csv('customers.csv', index=False)
```

```python
# Create product data

categories = ['Electronics','Clothing','Home','Beauty','Sports']

products = []
for i in range(1, 201):
    products.append([i,
        fake.word().capitalize(),
        random.choice(categories),
        round(random.uniform(10, 500), 2)])

products_df = pd.DataFrame(products, columns=['product_id','product_name','category','price'])

products_df.to_csv('products.csv', index=False)
```

```python
# Create Orders data

orders = []
for i in range(1, 5001):
    orders.append([i,
        random.randint(1,1000),
        fake.date_between(start_date='-2y', end_date='today'),
        random.choice(['Credit Card','Debit Card','UPI','COD']),
        random.choice(['Delivered','Delivered','Cancelled'])])

orders_df = pd.DataFrame(orders, columns=['order_id','customer_id','order_date','payment_method','order_status'])

orders_df.to_csv('orders.csv', index=False)
```

```python
# Create Order items data

order_items = []
order_item_id = 1

for order_id in range(1, 5001):
    for _ in range(random.randint(1,4)):
        order_items.append([order_item_id,
            order_id,
            random.randint(1,200),
            random.randint(1,5),
            random.choice([0,5,10,15,20])])
        order_item_id += 1

order_items_df = pd.DataFrame(order_items, columns=['order_item_id','order_id','product_id','quantity','discount'])

order_items_df.to_csv('order_items.csv', index=False)
```
